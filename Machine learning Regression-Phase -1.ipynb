```python
# 1. Data Collection:

# Import pandas libraries for data manupulation (tables,database) & data analysis:

import pandas as pd

# Create variable name to read or view data collected (.csv) file:

dataset=pd.read_csv("Salary_Data.csv")

# Call the data to read or view  data collected (.csv) file:

dataset
```

```python
# 2. Data Preprocessing:  

# Data is already prepocessed.
```

```python
# 3. Split Input & Output:

# To Split Input:

independent=dataset[["YearsExperience"]]

# To split Output:

dependent=dataset[["Salary"]]
```

```python
# To Call or view splited input data:
independent
```

```python
# To call or view splited output:

dependent
```

```python
# 4. To split Train set & Test set:

# sklearn- it trains your model in simple words:
  # a) Looks at your data → reads numbers in the table

  # b) Finds patterns → sees salary increases with experience

  # c) Learns relationships → builds a rule (a line) that connects experience to salary

from sklearn.model_selection import train_test_split

#  Break the data into two parts:
# x_train, y_train = for teaching the model
# x_test, y_test = for checking the model

# test_size=0.20 → use 20% of the data for testing
# random_state=40 → make the split the same every time

x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.20,random_state=40)
```

```python
# To view the splited x_train (80% of input) data:
x_train
```

```python
# To view the splited x_test (20% of input) data:
x_test
```

```python
# To view the splited y_train (80% of output) data:

y_train
```

```python
# To view the splited y_test (20% of output) data:

y_test
```

```python
# 5. Model creation part:

# sklearn is a library in Python for machine learning.

# linear_model contains algorithms for linear regression.

# LinearRegression is the class that implements linear regression.

from sklearn.linear_model import LinearRegression

# Create a LinearRegression model (this is just an empty model for now)

regressor=LinearRegression()

# Train the model using the training data

# x_train -> input features

# y_train -> target/output values

# The model learns the relationship between x_train and y_train

regressor.fit(x_train,y_train)
```

```python
# To get weight or slope:

# regressor.coef_ → Gives the weights 

weight=regressor.coef_
```

```python
# To store weight or slope:

weight
```

```python
# To get bias value where starts

# regressor.intercept_ gives you that starting point directly.

bias=regressor.intercept_
```

```python
# To call Bias:

bias
```

```python
# Use the trained model to make predictions on new data

# x_test -> the input features for which we want predictions

# regressor.predict() -> calculates the predicted target values using the learned weights and bias

# y_pred -> stores the predicted values

y_pred=regressor.predict(x_test)
```

```python
# To call or stores predicted values:

y_pred
```

```python
# 6. Evaluate the Metrics:

# Import the r2_score function from scikit-learn

# r2_score is used to measure to check how good the model is

from sklearn.metrics import r2_score

# calculate R² score by comparing actual salaries with predicted salaries

r_score=r2_score(y_test,y_pred)
```

```python
# To call or view good model:

r_score
```

```python
# 7. Save the good model:

# pickle is a Python library to save and load Python objects (like models, lists, dictionaries) 

import pickle

# filename - filename to save or load your model without typing the file name again.

# The .sav extension is a common choice for saved models

filename="finalized_new_model.sav"
```

```python
# Take the trained model and save it to the file so it can be loaded later.

# wb – save your object safely so you can use it later without any change.

pickle.dump(regressor,open(filename, 'wb'))
```

```python
# 7. Load the Saved new evaluated model:

# Load the saved model from the file

# 'rb' = read in binary mode

loaded_model=pickle.load(open("finalized_new_model.sav",'rb'))

# Use the loaded model to predict a value for new data

# [[15]] = input value (1 row, 1 feature)

result=loaded_model.predict([[15]])
```
